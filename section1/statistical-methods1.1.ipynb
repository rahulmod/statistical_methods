```python
import pandas as pd
from pandas import DataFrame
from sklearn.datasets import load_iris
iris_obj = load_iris()
iris_obj.feature_names
```

```python
iris_obj.target
```

```python
iris_obj.target_names
```

```python
iris = DataFrame(iris_obj.data, columns=iris_obj.feature_names, index=pd.Index([i for i in range(iris_obj.data.shape[0])])).join(
DataFrame(iris_obj.target, columns=pd.Index(["species"]), index=pd.Index([i for i in range(iris_obj.target.shape[0])])))
iris.species.replace({0: 'setosa', 1: 'versicolor', 2: 'verginica'}, inplace=True)
iris
```

```python
iris_grps = iris.groupby("species")
for name, data in iris_grps:
    print(name)
    print("----------------------\n\n")
    print(data.iloc[:, 0:4])
    print("\n\n\n")
```

```python
def range_stat(s):
    return s.max() - s.min()
iris.iloc[:, 0:4].apply(range_stat)
```

```python
#apply is for application of algorithm/function row wise
#aggregate is for application of algorith/function group wise; so first group by an dthen apply aggregate
iris_grps.aggregate(range_stat)
```
